### CoLA -> MRPC -> NER

### CoLA: 0(ungrammatical), 1(grammatical)



**cnt1** : indexes predicted correctly in the first time => 860

**cnt2** : indexes predicted correctly in the second time => 789

**cnt3** : indexes predicted correctly in the first time but predicted uncorrectly in the second time (forget) => 118
  - **cnt3_0** : how many label '0' are forgot? => 117
  - **cnt3_1** : how many label '1' are forgot? => 1

**cnt4** : indexes predicted correctly in the second time but predicted uncorrectly in the first time => 47
  - **cnt4_0** : how many label '0' are learned? => 0
  - **cnt4_1** : how many label '1' are learned? => 47

```
def accuracy(lists):
  cnt1= []
  cnt2 = []
  cnt3 = []
  cnt3_0 = []
  cnt3_1 = []
  cnt4 = []
  cnt4_0 = []
  cnt4_1 = []
  for i, seq in enumerate(lists):
    if lists[i][1][0] == lists[i][1][1]:
      cnt1.append(i)
    if lists[i][1][0] == lists[i][1][2]:
      cnt2.append(i)
  
  for i in cnt1:
    if i not in cnt2:
      cnt3.append(i)
      if lists[i][1][0] == '0':
        cnt3_0.append(i)
      elif lists[i][1][0] == '1':
        cnt3_1.append(i)

  for i in cnt2:
    if i not in cnt1:
      cnt4.append(i)
      if lists[i][1][0] == '0':
        cnt4_0.append(i)
      elif lists[i][1][0] == '1':
        cnt4_1.append(i)
  return len(cnt1), len(cnt2), len(cnt3), len(cnt3_0),len(cnt3_1), len(cnt4), len(cnt4_0),len(cnt4_1)
accuracy(lists)
```

```
def read_data_from_file_ner(data_dir):
  dic = defaultdict()
  labels = []
  preds = []
  with open(data_dir, encoding='utf-8') as f:
    for i,line in enumerate(f):
      # print(line)
      if i == 0:
        continue
      if line[0] == '\n':
        if len(labels) > 0:
          # print(key)
          # print(labels)
          # print(preds)
          dic[key] = [labels,preds]
          labels = []
          preds = []
      else:
        splits = line.split('\t')
        # print(splits)
        if splits[0].isdigit and splits[0] != '':
          key = splits[0]
        labels.append(splits[1])
        preds.append(splits[2][:-1])
  return dic
```

```
dic1 = read_data_from_file_ner('eval_results_ner1.txt')
dic2 = read_data_from_file_ner('eval_results_ner2.txt')

list1 = sorted(dic1.items(), key=lambda x:int(x[0]))
list2 = sorted(dic2.items(), key=lambda x:int(x[0]))
print(len(list1))
print(len(list2))
print(list2[-1])
```

```
cpm_dict = defaultdict()
for i, (item1,item2) in enumerate(zip(list1, list2)):
  assert item1[0] == item2[0]
  assert len(item1[1][0]) == len(item2[1][0])
  cpm_dict[item1[0]] = [item1[1][0], item1[1][1], item2[1][1]]
```

```
label_list = ["O", "B-MISC", "I-MISC", "B-PER", "I-PER", "B-ORG", "I-ORG", "B-LOC", "I-LOC"]
```

**cnt1** : how many times is each label correctly predicted in the first time? 

**cnt2** : how many times is each label correctly predicted in the second time? 

**cnt3** : how many times is each label correctly predicted in the first time but uncorrectly in the second time? 

**cnt4** : how many times is each label correctly predicted in the second time but uncorrectly in the first time? 

```
cnt1 = [0]*len(label_list)
cnt2 = [0]*len(label_list)
cnt3 = [0]*len(label_list)
cnt4 = [0]*len(label_list)
for item in cpm_dict.items():
  # print(item)
  for l1,l2,l3 in zip(item[1][0], item[1][1], item[1][2]):
    if l1 == l2:
      for i,label in enumerate(label_list):
        if l1 == label:
          cnt1[i] += 1
    if l1 == l3:
      for i,label in enumerate(label_list):
        if l1 == label:
          cnt2[i] += 1 
    if l1 == l2 and l1 != l3:
      for i,label in enumerate(label_list):
        if l1 == label:
          cnt3[i] += 1  
    if l1 == l3 and l1 != l2:
      for i,label in enumerate(label_list):
        if l1 == label:
          cnt4[i] += 1    

print(cnt1)  
print(cnt2) 
print(cnt3) 
print(cnt4) 
label_list = ["O", "B-MISC", "I-MISC", "B-PER", "I-PER", "B-ORG", "I-ORG", "B-LOC", "I-LOC"]
```
